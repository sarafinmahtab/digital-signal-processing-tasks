```python
#Task#03: Create 3 Sinusoidal signal y1, y2, y3. then combine them and plot it.

A = 4
fs = 60
f1 = 5
f2 = 10
f3 = 15
sample = 60

x = np.arange(sample)
y1 = A * np.sin(2 * np.pi * f1 * (x/fs))
y2 = A * np.sin(2 * np.pi * f2 * (x/fs))
y3 = A * np.sin(2 * np.pi * f3 * (x/fs))


pyplot.rcParams["figure.figsize"] = (16,8)

pyplot.subplot(221, frameon=True)
pyplot.stem(x,y1)
pyplot.plot(x,y1)
pyplot.title('1st')

pyplot.subplot(222, frameon=True)
pyplot.stem(x,y2)
pyplot.plot(x,y2)
pyplot.title('2nd')

pyplot.subplot(223, frameon=True)
pyplot.stem(x,y3)
pyplot.plot(x,y3)
pyplot.title('3rd')

y = y1+y2+y3
pyplot.subplot(224, frameon=True)
pyplot.stem(x,y)
pyplot.plot(x,y)
pyplot.title('Combined')

pyplot.xlabel("Amplitude")
pyplot.ylabel("Frequency")

pyplot.show()
```

```python
#Task#04: Plot the basic signals like unit impulse, unit samples, unit step, exponential, unit ramp (discussed in classes) etc.

pyplot.rcParams["figure.figsize"] = (16,16)

#signal_array = [-3,2,-4,1,0,3,2,1,-1,4]
limit = 4

x = np.arange(-limit+1,limit)

#unit impulse

impulse=[]
for i in range(-limit+1, limit):
    if i==0:
        impulse.append(1)
    else:
        impulse.append(0)

pyplot.subplot(221)
pyplot.axvline(0)
pyplot.axhline(0)
pyplot.xlim(-limit, limit)
pyplot.ylim(-limit, limit)
pyplot.stem(x, impulse)
pyplot.title('Unit Sample Sequence')

#unit step signal

step=[]
for i in range(-limit,-1):
    step.append(0)
for i in range(limit):
    step.append(1)

pyplot.subplot(222)
pyplot.axvline(0)
pyplot.axhline(0)
pyplot.xlim(-limit, limit)
pyplot.ylim(-limit, limit)
pyplot.stem(x,step)
pyplot.title('Unit Step Signal')

#unit ramp

ramp = []

for i in range(-limit,-1):
    ramp.append(0)
for i in range(limit):
    ramp.append(i)

pyplot.subplot(223)
pyplot.axvline(0)
pyplot.axhline(0)
pyplot.xlim(-limit, limit)
pyplot.ylim(-limit, limit)
pyplot.stem(x,ramp)
pyplot.title('Unit Ramp Signal')

#unit exponential

expo = []
power = 3
for i in range(-limit+1,limit):
    expo.append(power**i)

pyplot.subplot(224)
pyplot.axvline(0)
pyplot.axhline(0)
pyplot.xlim(-limit, limit)
pyplot.ylim(-limit*7, limit*7)
pyplot.stem(x,expo)
pyplot.title('Exponential Signal')
```

```python
#Task#06: Sampling Theorem Experiment.

#The sampling theorem states that a continuous signal x(t) 
#bandlimited to B Hz can be recovered from its samples x[n] = x(n*T), 
#where n is an integer, if T is greater than or equal to 1/(2B) 
#without loss of any information. And we call 2B the Nyquist rate.

from numpy import linspace,cos,pi,ceil,floor,arange
from pylab import plot,show,axis

# Signal sampling
B = 10;
tmin = -0.5;
tmax = 0.5;
t = linspace(tmin, tmax, 1000);
x = cos(2*pi*B*t);
plot(t, x)

# using Nyquest rate
T = 1/(2*B);
nmin = ceil(tmin / T);
nmax = floor(tmax / T);
n = arange(nmin,nmax);

x1 = cos(2*pi*f*n*T);
plot(n*T, x1, color='green', marker='o', linestyle='dashed', linewidth=2, markersize=12)

# undersampling
T = 1/15; # 2*B > 15
nmin = ceil(tmin / T);
nmax = floor(tmax / T);
n = arange(nmin,nmax);
x2 = cos(2*pi*f*n*T);
plot(n*T, x2, '-r.',markersize=5)

axis([-0.5, 0.5, -1.5, 1.5])
show()
```

```python
#Task#07: Implement some basic systems like Adder, Multiplier, Delay, Advanced, Fold etc.

signal = [-3,2,-4,1,0,3,2,1,-1]
print('Main Signal: ', signal)

limit = 4

x = np.arange(-limit,len(signal)-limit)

pyplot.rcParams["figure.figsize"] = (10,10)

pyplot.subplot(411)
pyplot.axvline(0)
pyplot.axhline(0)
pyplot.xlim(-10, 10)
pyplot.ylim(-10, 10)
pyplot.stem(x,signal)
pyplot.title('Main signal')



delay_unit = 4
delay = []

for i in range(0,delay_unit):
    delay.append(0)
for i in range(0,len(signal)):
    delay.append(signal[i])

x = np.arange(-limit,len(delay)-limit)

pyplot.subplot(412)
pyplot.axvline(0)
pyplot.axhline(0)
pyplot.xlim(-10, 10)
pyplot.ylim(-10, 10)
pyplot.stem(x,delay)
pyplot.title('Delay signal')



advance_unit = 4
advance = []

for i in range(0,len(signal)):
    advance.append(signal[i])
for i in range(0,advance_unit):
    advance.append(0)


x = np.arange(-(limit+advance_unit),len(advance)-(limit+advance_unit))

pyplot.subplot(413)
pyplot.axvline(0)
pyplot.axhline(0)
pyplot.xlim(-10, 10)
pyplot.ylim(-10, 10)
pyplot.stem(x,advance)
pyplot.title('Advance signal')



fold = []
for i in reversed(signal):
    fold.append(i)

x = np.arange(-limit,len(fold)-limit)

pyplot.rcParams["figure.figsize"] = (10,10)

pyplot.subplot(414)
pyplot.axvline(0)
pyplot.axhline(0)
pyplot.xlim(-10, 10)
pyplot.ylim(-10, 10)
pyplot.stem(x,fold)
pyplot.title('Fold signal')


# Adder and Multipler of Main and Fold signal 
adder = []
multiplier = []
for i in range(0, len(signal)):
    adder.append(signal[i]+fold[i])
    multiplier.append(signal[i]*fold[i])

print('Adder: ', adder)
print('Multiplier: ', multiplier)
```

```python
#Task#08: Implement Moving average filter, Median Filter, Accumulator etc.

import scipy

pyplot.rcParams["figure.figsize"] = (16,8)

fs = 90
f1 = 5
f2 = 10
f3 = 15
sample = 100
A = 5

x = np.arange(sample)

y1 = A * np.sin(2 * np.pi * f1 * (x/fs))
y2 = A * np.sin(2 * np.pi * f2 * (x/fs))
y3 = A * np.sin(2 * np.pi * f3 * (x/fs))

y = y1+y2+y3
pyplot.subplot(411)
pyplot.stem(x, y)
pyplot.plot(x, y)
pyplot.title('Main Signal')


Y = y1+y2+y3

#Moving average filter
point = 3
for i in range(int(point/2)):
    Y[i]=0
    Y[i]=y[i]

for i in range(int(point/2),int(len(y)-point/2)):
    Y[i]=0
    for j in range(-int(point/2),int(point/2+1)):
        Y[i] =Y[i]+y[i+j]
    Y[i]/=point

pyplot.subplot(412)
pyplot.stem(x, Y)
pyplot.plot(x, Y)
pyplot.title('Moving average filter')


#Median filter
point = 3
for i in range(int(point/2)):
    Y[i]=y[i]
    
for i in range(int(point/2),len(y)-int(point/2)):
    Y[i]=0
    a=[]
    for j in range(-int(point/2),int(point/2+1)):
        a.append(y[i+j])
    a.sort()
    Y[i]=a[int(point/2)]

pyplot.subplot(413)
pyplot.stem(x, Y)
pyplot.plot(x, Y)
pyplot.title('Median filter')


#Accumulator fiter
for i in range(len(y)):
    Y[i]=0
    for j in range(i+1):
        Y[i]=Y[i]+y[j]

pyplot.subplot(414)
pyplot.stem(x, Y)
pyplot.plot(x, Y)
pyplot.title('Accumulator filter')
```

```python
#Task#09: Convolution

# Convolution implementation
x= [1,-2,2,1]
h= [1,2,1]
xlen=len(x)
hlen=len(h)

X=[]
H=[]
Y=[]

for i in range(xlen+hlen-1):
    if i<xlen:
        X.append(x[i])
    else:
        X.append(0)
    
    if i<hlen:
        H.append(h[i])
    else:
        H.append(0)
    Y.append(0)


for i in range(xlen+hlen-1):
    for j in range(xlen+hlen-1):
        Y[i] = Y[i]+X[j]*H[i-j];


# Convolution function
y=np.convolve(x,h)

# Comparison
print(Y, y)
```

```python
#Task#14: circular convolution program

x=[2,1,2,1]
h=[1,2,3,4]

fold=[]
y =[]
Y=[]

# Folding
fold.append(x[0])
for i in range(len(x)-1,0,-1):
    fold.append(x[i])


for i in range(len(x)):
    d=[]
    # Multiplication
    for j in range(len(fold)):
        d.append(fold[j]*h[j])
    # Summation
    sum=0
    for j in range(len(d)):
        sum=sum+d[j]
    
    y.append(sum)
    
    shift=[]
    
    # Shifting anti-clock wise by 1 unit
    shift.append(fold[len(fold)-1])
    for j in range(len(fold)-1):
        shift.append(fold[j])
    
    fold=shift
print(y)

Y = np.convolve(x,h)
print(Y)
```

```python

from pylab import *
import scipy.signal as signal

# Frequency and Phase response
def mfreqz(b,a=1):
    w,h = signal.freqz(b,a)
    h_dB = 20 * log10 (abs(h))
    subplot(211)
    plot(w/max(w),h_dB)
    ylim(-150, 5)
    ylabel('Magnitude (db)')
    xlabel(r'Normalized Frequency (x$\pi$rad/sample)')
    title(r'Frequency response')
    subplot(212)
    h_Phase = unwrap(arctan2(imag(h),real(h)))
    plot(w/max(w),h_Phase)
    ylabel('Phase (radians)')
    xlabel(r'Normalized Frequency (x$\pi$rad/sample)')
    title(r'Phase response')
    subplots_adjust(hspace=0.5)

# Step and Impulse response
def impz(b,a=1):
    l = len(b)
    impulse = repeat(0.,l); impulse[0] =1.
    x = arange(0,l)
    response = signal.lfilter(b,a,impulse)
    subplot(211)
    stem(x, response)
    ylabel('Amplitude')
    xlabel(r'n (samples)')
    title(r'Impulse response')
    subplot(212)
    step = cumsum(response)
    stem(x, step)
    ylabel('Amplitude')
    xlabel(r'n (samples)')
    title(r'Step response')
    subplots_adjust(hspace=0.5)
```

```python
#Task#17: FIR Filter Low Pass

n = 51
a = signal.firwin(n, cutoff = 0.3, window = "hamming")
#Frequency and phase response
mfreqz(a)
show()
#Impulse and step response
figure(2)
impz(a)
show()
```

```python
#Task#18: FIR Filter High Pass

n = 101
a = signal.firwin(n, cutoff = 0.3, window = "hanning", pass_zero=False)
#Frequency and phase response
mfreqz(a)
show()
#Impulse and step response
figure(2)
impz(a)
show()
```
